```python
import numpy as np
import pandas as pd

import experiments
```

```python
random_states = np.array([
    2587, 8074, 3192, 8357, 9981, 5260, 3867, 5358, 4595, 2286, 2813,
    7749, 7645, 7278, 3523, 1682, 4319, 3399,  482, 8889, 9934, 6505,
    7738, 3555, 7250, 2520, 7085, 5000,  288, 7098])

result = experiments.experiment_network_sir(
    epochs=60,
    random_states=random_states,
    n_jobs=-1)
```

```python
results = pd.concat(result, axis=0).reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

def plot_sir_model_result(results):
    grouped = results.groupby('index')
    means = grouped.apply(np.mean)
    stds = grouped.apply(np.std)
    COLORS = {
        'susceptible': 'blue',
        'infected': 'red',
        'recovered': 'orange'
    }
    LABELS = {
        'susceptible': 'Suscetível',
        'infected': 'Infectado',
        'recovered': 'Recuperado'
    }
    plt.figure(figsize=(10, 5))
    for ltype in ['susceptible', 'infected', 'recovered']:
        plt.fill_between(
            means.index,
            means[ltype] - 1.96 * stds[ltype],
            means[ltype] + 1.96 * stds[ltype],
            color='black')
        plt.plot(means[ltype], c=COLORS[ltype], label=LABELS[ltype])
    ax = plt.gca()
    ax.set_ylim((0, 5000000))
    vals = ax.get_yticks()
    plt.yticks(np.arange(0, vals.ptp(), step=500000))
    vals = ax.get_yticks()
    ax.set_yticklabels(['{:.1f}MM'.format(x / 1000000) for x in vals])
    plt.xticks(np.where(means.index % 6 == 0)[0])
    plt.title('SIR Model. Tx infecção: 15%, Tx Recuperação: 1%, 500 infectados iniciais')
    plt.legend()
    plt.show()
    
plot_sir_model_result(results)
```

```python
results.to_csv('resultados.csv', index=False)
```
